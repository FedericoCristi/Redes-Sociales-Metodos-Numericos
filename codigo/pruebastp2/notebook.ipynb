Grafo de centralidad:
corrido con 10000 iteraciones, 0.00001 tolerancia

```python
import networkx as nx
import random

#cargar grafo
df = np.genfromtxt("../datos/karateclub_matriz.txt")
grafo = nx.from_numpy_array(df)

#cargar valores de centralidad
vec = [ 0.355923, 0.266372, 0.317432, 0.211729, 0.076305, 0.079851, 0.079851, 0.171514, 0.227549, 0.102718, 0.076305, 0.053101, 0.084601, 0.227009, 0.101323, 0.101323, 0.023730, 0.092791, 0.101323, 0.148285, 0.101323, 0.092791, 0.101323, 0.149836, 0.056857, 0.058958, 0.075441, 0.133386, 0.131009, 0.134680, 0.174742, 0.191136, 0.307634, 0.371973 ] 

nodes = [i+1 for i in range(len(vec))]

grafo = nx.relabel_nodes(grafo, lambda x: x+1)
grafo.nodes(data=True)

random.seed(10)

options = {"edgecolors": "tab:gray", "node_size": 800, "alpha": 0.9}
nx.draw_kamada_kawai(grafo, with_labels=True, cmap=plt.cm.summer, **options)

plt.show()

vec_names = ["%.2f"%item for item in vec]
for i in range(0,len(vec_names)):
    add = "\u200b"
    for j in range(0, i):
        add = add + "\u200b"
    vec_names[i] = str(vec_names[i]) + add

dicc = dict(zip(grafo.nodes, vec_names))
grafo2 = nx.relabel_nodes(grafo, dicc)

nx.draw_kamada_kawai(grafo2, with_labels=True, node_color=vec, cmap=plt.cm.summer, **options)

plt.show()
```
